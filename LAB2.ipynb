```python
class Employee:
    def __init__(self, first_name, last_name, salary, experience):
        self.first_name = first_name
        self.last_name = last_name
        self.salary = salary
        self.experience = experience

class Developer(Employee):
    pass

class Designer(Employee):
    def __init__(self, first_name, last_name, salary, experience, efficiency):
        super().__init__(first_name, last_name, salary, experience)
        self.efficiency = efficiency

class Manager(Employee):
    def __init__(self, first_name, last_name, salary, experience, team):
        super().__init__(first_name, last_name, salary, experience)
        self.team = team


class Employee:
    def __init__(self, first_name, last_name, base_salary, experience):
        self.first_name = first_name
        self.last_name = last_name
        self.base_salary = base_salary
        self.experience = experience
        
    def countedSalary(self):
        if self.experience > 5:
            return self.base_salary * 1.2 + 500
        elif self.experience > 2:
            return self.base_salary + 200
        else:
            return self.base_salary

class Developer(Employee):
    pass

class Designer(Employee):
    def __init__(self, first_name, last_name, base_salary, experience, eff_coeff):
        super().__init__(first_name, last_name, base_salary, experience)
        self.eff_coeff = eff_coeff
    
    def countedSalary(self):
        return super().countedSalary() * self.eff_coeff

class Manager(Employee):
    def __init__(self, first_name, last_name, base_salary, experience, team):
        super().__init__(first_name, last_name, base_salary, experience)
        self.team = team
        
    def countedSalary(self):
        base_salary = super().countedSalary()
        bonus = 0
        if len(self.team) > 10:
            bonus += 300
        elif len(self.team) > 5:
            bonus += 200
        num_developers = sum(isinstance(emp, Developer) for emp in self.team)
        num_designers = sum(isinstance(emp, Designer) for emp in self.team)
        if num_developers > num_designers:
            bonus += base_salary * 0.1
        return base_salary + bonus

class Department:
    def __init__(self, managers):
        self.managers = managers
        
    def giveSalary(self):
        for manager in self.managers:
            for emp in manager.team:
                salary = emp.countedSalary()
                print(f"{emp.first_name} {emp.last_name} отримав {salary} грошей.")



dev1 = Developer("John", "Doe", 3000, 4)
dev2 = Developer("Jane", "Doe", 4000, 6)
des1 = Designer("Bob", "Smith", 3500, 3, 0.8)
des2 = Designer("Alice", "Johnson", 4500, 7, 0.9)
man1 = Manager("Tom", "Jones", 5000, 8, [dev1, des1])
man2 = Manager("Sue", "Davis", 5500, 10, [dev2, des2, man1])


dept = Department([man1, man2])
dept.giveSalary()
```
